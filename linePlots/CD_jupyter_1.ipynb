```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load birth data from CSV files
birth_data = pd.read_csv('data/birthData.csv')
birth_estimate = "Estimates, 1950 - 2020: Annually interpolated demographic indicators - Births (thousands)"
```

```python
#Select rows about afghanistan, africa, and albania
AFG = birth_data[birth_data.Entity == 'Afghanistan']
AFR = birth_data[birth_data.Entity == 'Africa']
ALB = birth_data[birth_data.Entity == 'Albania']
```

```python
#merge into a single dataframe
birthsA = pd.concat([AFG, AFR, ALB])
```

```python
afgBirths = np.array(AFG.iloc[:, 3])
afrBirths = np.array(AFR.iloc[:, 3])
albBirths = np.array(ALB.iloc[:, 3])
```

```python
AFG.plot(x = "Year", y = birth_estimate, kind = "line")
```

```python
ax = plt.gca()

AFG.plot(x = "Year", y = birth_estimate, kind = "line", label = "Afghanistan", ax = ax)
ALB.plot(x = "Year", y = birth_estimate, kind = "line", label = "Albania", ax = ax)
AFR.plot(x = "Year", y = birth_estimate, kind = "line", label = "Africa", ax = ax)
```

```python
fig, ax = plt.subplots()
ax.plot(AFG.Year, afgBirths, label="Afghanistan")
ax.plot(ALB.Year, albBirths, label="Albania")
ax.plot(AFR.Year, afrBirths, label="Africa")
ax.set_xlabel('Year')
ax.set_ylabel('Estimated Births in Thousands')
ax.set_title("Births across 3 countries")
```
